Par défaut, *x* est une variable symbolique (vous pouvez vérifier en faisant un `type(x)` sans définir la variable), permettant de définir des expression *symboliques*. Si vous déclarez un PolynomialRing, la variable n'est plus symbolique: il s'agit d'un polynôme (pas le même objet => pas les même méthodes disponibles).

Il ne faut pas oublier d'indiquer expressement le multiplications quand on a une forme factorisée, sinon vous interpretez le premier élement en le second élément :
* (x^2 + 5)(x^2 - 7) = x^4 - 14\*x^2 + 54. Ici on a définit une fonction *f(x) = x^2 + 5* et on l'évalue en *x^2 - 7*
* (x^2 + 5) __\*__ (x^2 - 7) = x^4 - 2\*x^2 - 35

```python
# Exercice 1 - 1
pol = (x**2+5)*(x**2-7)
print("pol(x) = ", pol.expand())
print("pol(5) = ", pol(5))
print("roots: ", pol.roots())
```

```python
# Exercice 1 - 2
def fac_it(n):
    res = 1
    for i in range(2, n+1):
        res *= i
    return res

def fac_rec(n):
    if n == 1: 
        return 1
    return n * fac_rec(n-1)

print("Test on 20 iterations:")
print("\tfac_it:", timeit("fac_it(20)"))
print("\tfac_rec:", timeit("fac_rec(20)"))
print("\tfactorial:", timeit("factorial(20)"))

print("Test on 250 iterations:")
print("\tfac_it:", timeit("fac_it(250)"))
print("\tfac_rec:", timeit("fac_rec(250)"))
print("\tfactorial:", timeit("factorial(250)"))
```

```python
# Exercice 3 - 4
def extended_euclide(a, b):
    r_old, r = a, b
    s_old, s = 1, 0
    t_old, t = 0, 1
    
    while r != 0:
        q = r_old // r
        r_old, r = r, r_old - q * r
        s_old, s = s, s_old - q * s
        t_old, t = t, t_old - q * t
    
    gcd, bezout = r_old, (s_old, t_old)
    
    return gcd, bezout

g, b = extended_euclide(252, 90)
print("252*{} + 90*{} = {}".format(b[0], b[1], g))
```

```python
# Exercice 6

from binascii import unhexlify

def rsa_key_gen(bitlen, e = 65537):
    p = random_prime(2**(bitlen//2-1), 2**(bitlen//2))
    q = random_prime(2**(bitlen//2-1), 2**(bitlen//2))
    n = p*q
    phi = (p-1)*(q-1)
    d = e.inverse_mod(phi)

    return (n, e), (n, d, p, q)

def decode(x):
    h = hex(x)[2:]
    if len(h) % 2 == 1:
        h = '0' + h
    return unhexlify(h)

def char_to_bytestr(c):
    x = hex(ord(c))[2:]
    return x if len(x) == 2 else '0'+x 

def encode(m):
    i = ''.join([char_to_bytestr(c) for c in m])
    return int(i, 16)

def encrypt(m, e, n):
    m_int = encode(m)
    if m_int >= n:
        return
    return pow(m_int, e, n)

def decrypt(c, d, n):
    pt = pow(c, d, n)
    return decode(pt)

pub, priv = rsa_key_gen(2048)
m = "test message"
ct = encrypt(m, pub[1], pub[0])
pt = decrypt(ct, priv[1], priv[0])
print("ct =", hex(ct))
print("pt =", pt)
```
